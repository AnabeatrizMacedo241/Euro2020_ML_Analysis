#### 3.2 Predict passing performance of the players
We then decide to help the coach create plays based on their player's passing skills during the Euro 2020 using our SVM Machine Learning model.

We used every single match data from all of the 51 games in order to do that.

```python
# Getting all Matches from UEFA Men's Euro 2020 to trin the Passing evaulation model
from statsbombpy import sb
import pandas as pd
import numpy as np
df = sb.matches(competition_id=55, season_id=43)
match_ids = df['match_id'].unique().tolist()
match_ids = match_ids[:5] #make dataset smaller for testing purposes
first_match_id = match_ids[0]
match_events = sb.events(match_id=first_match_id)
for i in match_ids[1:]:
    append = sb.events(match_id=i)
    match_events = pd.concat([match_events, append])
#match_events.to_csv('all_matches_data.csv')
```

```python
#Data exploration to filter by Pass and get the locations of the passers and recievers
match_events = match_events[match_events['type'] == 'Pass']
match_events[['location_x', 'location_y']] = match_events['location'].apply(pd.Series)
match_events[['pass_end_location_x', 'pass_end_location_y']] = match_events['pass_end_location'].apply(pd.Series)

relevant_columns = ['player', 'location_x', 'location_y', 'timestamp', 'team', 'pass_recipient',
                    'pass_height', 'pass_outcome', 'pass_body_part', 'pass_type', 'pass_length',
                    'pass_end_location_x', "pass_end_location_y", 'duration', 'pass_angle']
match_events[relevant_columns]

match_events[['pass_aerial_won', 'pass_angle', 'pass_assisted_shot_id',
       'pass_body_part', 'pass_cross', 'pass_cut_back', 'pass_deflected',
       'pass_end_location', 'pass_goal_assist',
       'pass_inswinging', 'pass_length', 'pass_miscommunication',
       'pass_no_touch', 'pass_outcome', 'pass_outswinging', 'pass_recipient',
       'pass_shot_assist', 'pass_switch',
       'pass_through_ball']]
```

```python
#Split locations into x and y components
match_events[['location_x', 'location_y']] = match_events['location'].apply(pd.Series)
match_events[['pass_end_location_x', 'pass_end_location_y']] = match_events['pass_end_location'].apply(pd.Series)

#Split the shot_end_locations into x,y and z - some z missing
match_events['shot_end_location_x'], match_events['shot_end_location_y'], \
    match_events['shot_end_location_z'] = np.nan, np.nan, np.nan
```

```python
#Cean up and choose the columns wanted for the model
events_df = match_events.copy()
events_df = events_df[[
    'index', 'id', 'duration', 'match_id', 'minute', 'pass_angle', 'pass_height', 'pass_length', 'pass_outcome',
    'pass_recipient', 'pass_shot_assist', 'period', 'play_pattern', 'player', 'player_id', 'position', 'possession',
    'possession_team', 'possession_team_id', 'shot_type', 'shot_outcome', 'team', 'timestamp', 'type', 'location_x',
    'location_y', 'pass_end_location_x', 'pass_end_location_y', 'shot_end_location_x', 'shot_end_location_y',
    'shot_end_location_z'
]]
events_df.loc[:, 'pass_outcome'] = np.where(events_df['pass_outcome'].isnull(), 'Complete', events_df['pass_outcome'])
events_df = events_df[~events_df['pass_outcome'].isin(['Injury Clearance', 'Unknown'])]
pd.options.display.max_columns = None
events_df.head()
```

**Given factors, we want to calculate the probability of the pass completion by taking its average for each completed pass from the and then calculate the average of the completed pass predicted.**

```python
import sklearn
from sklearn import svm, preprocessing

pass_outcome_class_dict = {'Incomplete': 0, 'Out': 0, 'Pass Offside':0 ,'Complete':1}
events_df['pass_outcome'] = events_df['pass_outcome'].map(pass_outcome_class_dict)
events_df = sklearn.utils.shuffle(events_df)
```

```python
test_size = 25 #Use value 200 for full sample of Men Euro which will have approx 70k rows

df_train = events_df.copy()
df_train.drop(df_train.tail(test_size).index, inplace=True)

X = df_train[['duration', 'pass_angle', 'pass_length', 'location_x',
    'location_y', 'pass_end_location_x', 'pass_end_location_y']].values
X = preprocessing.scale(X)
y = df_train['pass_outcome'].values
X_train = X[:]
y_train = y[:]

df_test = events_df.copy()
df_test = df_test.tail(test_size)
X_te = df_test[['duration', 'pass_angle', 'pass_length', 'location_x',
    'location_y', 'pass_end_location_x', 'pass_end_location_y']].values
X_te = preprocessing.scale(X_te)
y_te = df_test['pass_outcome'].values
X_test = X_te[:]
y_test = y_te[:]

clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
#Checking how well the model is
print(f'R squared: {clf.score(X_test, y_test)}') #gives r squared
#calculate f1 score
from sklearn.metrics import classification_report
print(classification_report(y_train, clf.predict(X_train)))
```

After creating the model, we want to classify the player's passing skills based on the two probabilities.

- If the actual pass completion has a higher value than the predicted pass by our SVM model, the player will be classified as `Outperformed`.
- If the predicted probability is higher than the actual, the player will be classified as `Underperformed`.
- If the probabilities are equal, the player will be classified as `Perform to expectations`.

```python
final_df = df_train.copy()
#predicted_pass_completion is the proportion of passes that are predicted to be completed.
final_df['predicted_pass_completion'] = clf.predict(X_train)
final_df = final_df.groupby(['player'])[['predicted_pass_completion','pass_outcome']].mean()
final_df.rename(columns={'pass_outcome': 'actual_pass_completion'}, inplace=True)
#actual_pass_completion is the proportion of passes that are actually completed
#pass_potential compares the predicted pass outcome vs actual pass outcome
#a player tagged as "Outperform" has been completing more passes than expected given the predicted pass completion
final_df['pass_potential'] = np.where(final_df['predicted_pass_completion'] < final_df['actual_pass_completion'], 'Outperformed', '0')
final_df['pass_potential'] = np.where(final_df['predicted_pass_completion'] > final_df['actual_pass_completion'], 'Underperformed', final_df['pass_potential'])
final_df['pass_potential'] = np.where(final_df['predicted_pass_completion'] == final_df['actual_pass_completion'], 'Perform to expectations', final_df['pass_potential'])
```

```python
df2 = final_df.reset_index()
df2
```

```python
def PassPrediction(player_name):
    player = df2[df2['player']==player_name]
    return player
PassPrediction('Harry Kane')
```

**We can conclude that even though Harry Kane, as seen before, did have a good offensive performance and grade, he did not performed to expectations with his passing skills. Maybe he is better off as a scorer rather that creating plays**.

```python
def PlayerShotChart(player):
    shot = events_df[(events_df['type'] == 'Shot')]
    shot = shot[(shot['player'] == player)]
    goal = events_df[(events_df['shot_outcome'] == 'Goal')]
    goal = goal[(goal['player'] == player)]
    name = shot.player.to_list()
    fig, ax = plt.subplots(figsize=(13,8.5))
    fig.set_facecolor('#22312b')
    ax.patch.set_facecolor('#22312b')

    #The statsbomb pitch from mplsoccer
    pitch = Pitch(pitch_type='statsbomb',
                  pitch_color='#22312b', line_color='#c7d5cc',
                  )

    pitch.draw(ax=ax)

    #I invert the axis to make it so I am viewing it how I want
    plt.gca().invert_yaxis()

    #plot the points, you can use a for loop to plot the different outcomes if you want
    plt.scatter(shot['x'],shot['y'], s=100,c='blue',alpha=.7)
    plt.scatter(goal['x'],goal['y'], s=100,c='red',alpha=.7)

    s=f'{player} Shot Charts'
    fig.text(s=s,
            x=.3,y=.86,
            fontfamily='Andale Mono',
            fontsize=30,
            color='yellow'

    )
    total_shots = len(shot)
    total_goals = len(goal)

    fig.text(s=f'Total Shots: {total_shots}',
            x=.20, y =.80, fontsize=14,fontfamily='Andale Mono',color='white')
    fig.text(s=f'Total Goals: {total_goals}',
            x=.20, y =.77, fontsize=14,fontfamily='Andale Mono',color='white')
    
    player_shot = (events_df.player == player) & (events_df.type == 'Shot')
    df_shot = events_df.loc[player_shot, ['x', 'y']]
    #setup pitch
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2,
              pitch_color='#22312b', line_color='#efefef')
    # draw
    fig, ax = pitch.draw(figsize=(10.6,8.5))
    fig.set_facecolor('#22312b')
    bin_statistic = pitch.bin_statistic(df_shot.x, df_shot.y, statistic='count', bins=(25, 25))
    bin_statistic['statistic'] = gaussian_filter(bin_statistic['statistic'], 1)
    pcm = pitch.heatmap(bin_statistic, ax=ax, cmap='hot', edgecolors='#22312b')
    # Add the colorbar and format off-white
    cbar = fig.colorbar(pcm, ax=ax, shrink=0.6)
    cbar.outline.set_edgecolor('#efefef')
    cbar.ax.yaxis.set_tick_params(color='#efefef')
    ticks = plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='#efefef')    
    
    path_eff = [path_effects.Stroke(linewidth=1.5, foreground='black'),
            path_effects.Normal()]
    
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2, pitch_color='#22312b')
    fig, ax = pitch.draw(figsize=(9.7,8.5))
    fig.set_facecolor('#22312b')
    bin_statistic = pitch.bin_statistic(df_shot.x, df_shot.y, statistic='count', bins=(6, 5), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, cmap='Reds', edgecolor='#22312b')
    labels = pitch.label_heatmap(bin_statistic, color='#f4edf0', fontsize=18,
                             ax=ax, ha='center', va='center',
                             str_format='{:.0%}', path_effects=path_eff)
PlayerShotChart('Harry Kane')
```

```python
def PlayerPasses(player):

    rcParams['text.color'] = '#c7d5cc'  # set the default text color
    player_pass = (events_df.player == player) & (events_df.type == 'Pass')
    df_pass = events_df.loc[player_pass, ['x', 'y', 'end_x', 'end_y', 'pass_outcome']]
    player_pass_complete = df_pass.pass_outcome.isnull()
    
    # Set up the pitch
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
    fig, ax = pitch.draw(figsize=(10.6,8.5), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('#22312b')

    # Plot the completed passes
    pitch.arrows(df_pass[player_pass_complete].x, df_pass[player_pass_complete].y,
             df_pass[player_pass_complete].end_x, df_pass[player_pass_complete].end_y, width=2,
             headwidth=10, headlength=10, color='#ad993c', ax=ax, label='completed passes')

    # Plot the other passes
    pitch.arrows(df_pass[~player_pass_complete].x, df_pass[~player_pass_complete].y,
             df_pass[~player_pass_complete].end_x, df_pass[~player_pass_complete].end_y, width=2,
             headwidth=6, headlength=5, headaxislength=12,
             color='#ba4f45', ax=ax, label='other passes')

    # Set up the legend
    ax.legend(facecolor='#22312b', handlelength=5, edgecolor='None', fontsize=20, loc='lower left')

    # Set the title
    ax_title = ax.set_title(f'{player} passes', fontsize=30)
    
    p_len_player_pass = sum((events_df.player == player) & (events_df.type == 'Pass'))
    p_incomplete_pass = sum((events_df.player == player) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Incomplete'))
    p_injcl_pass = sum((events_df.player == player) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Injury Clearance'))
    p_out_pass = sum((events_df.player == player) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Out'))
    p_offside_pass = sum((events_df.player == player) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Pass Offside'))
    p_other_pass = sum((events_df.player == player) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Unknown'))
    
    p_incomplete_pass_all = p_incomplete_pass + p_injcl_pass + p_out_pass + p_offside_pass + p_other_pass
    p_complete_pass = p_len_player_pass - p_incomplete_pass_all
    
    textstr = '\n'.join((
    f'Complete Passes = {round((p_complete_pass/p_len_player_pass)*100)} %',
    f'Incomplete Passes = {round((p_incomplete_pass_all/p_len_player_pass)*100) } %',
    ))
    
    props = dict(boxstyle='round', facecolor='#7a7d7c', alpha=0.9)
    
    plt.text(0.05, 0.95, textstr, transform= ax.transAxes, fontsize=15,
        verticalalignment='top', bbox=props)
    
    #################
    
    flamingo_cmap = LinearSegmentedColormap.from_list("Flamingo - 100 colors",
                                                  ['#e3aca7', '#c03a1d'], N=100)
    
    pearl_earring_cmap = LinearSegmentedColormap.from_list("Pearl Earring - 10 colors",
                                                       ['#15242e', '#4393c4'], N=10)
    
    path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]
    
    # fontmanager for google font (robotto)
    robotto_regular = FontManager()
    
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2, pitch_color='#22312b')
    fig, axs = pitch.grid(title_height=0.08, title_space=0,
                      axis=False,
                      grid_height=0.84)
    fig.set_facecolor('#22312b')

    # heatmap and labels
    bin_statistic = pitch.bin_statistic_positional(df_pass.x, df_pass.y, statistic='count',
                                               positional='full', normalize=True)
    pitch.heatmap_positional(bin_statistic, ax=axs['pitch'],
                         cmap=flamingo_cmap, edgecolors='#22312b')
    labels = pitch.label_heatmap(bin_statistic, color='#f4edf0', fontsize=18,
                             ax=axs['pitch'], ha='center', va='center',
                             str_format='{:.0%}', path_effects=path_eff)

    # title
    axs['title'].text(0.5, 0.5, f'{player} passes', color='#dee6ea',
                  va='center', ha='center', path_effects=path_eff,
                  fontproperties=robotto_regular.prop, fontsize=25)
PlayerPasses('Harry Kane')
```

```python
def TeamShotChart(team):
    shot = events_df[(events_df['type'] == 'Shot')]
    shot = shot[(shot['team'] == team)]
    goal = events_df[(events_df['shot_outcome'] == 'Goal')]
    goal = goal[(goal['team'] == team)]
    name = shot.team.to_list()
    fig, ax = plt.subplots(figsize=(13,8.5))
    fig.set_facecolor('#22312b')
    ax.patch.set_facecolor('#22312b')
    #The statsbomb pitch from mplsoccer
    pitch = Pitch(pitch_type='statsbomb',
                  pitch_color='#22312b', line_color='#c7d5cc',
                  )
    pitch.draw(ax=ax)
    #I invert the axis to make it so I am viewing it how I want
    plt.gca().invert_yaxis()
    #Plot the points, you can use a for loop to plot the different outcomes if you want
    plt.scatter(shot['x'],shot['y'], s=100,c='blue',alpha=.7)
    plt.scatter(goal['x'],goal['y'], s=100,c='red',alpha=.7)
    s=f'{team} Shot Charts'
    fig.text(s=s,
            x=.325,y=.86,
            fontfamily='Andale Mono',
            fontsize=30,
            color='yellow'

    )
    total_shots = len(shot)
    total_goals = len(goal)
    fig.text(s=f'Total Shots: {total_shots}',
            x=.20, y =.80, fontsize=14,fontfamily='Andale Mono',color='white')
    fig.text(s=f'Total Goals: {total_goals}',
            x=.20, y =.77, fontsize=14,fontfamily='Andale Mono',color='white')
    team_shot = (events_df.team == team) & (events_df.type == 'Shot')
    df_team_shot = events_df.loc[team_shot, ['x', 'y']]
    #Setup pitch
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2,
              pitch_color='#22312b', line_color='#efefef')
    #Draw
    fig, ax = pitch.draw(figsize=(10.6,8.5))
    fig.set_facecolor('#22312b')
    bin_statistic = pitch.bin_statistic(df_team_shot.x, df_team_shot.y, statistic='count', bins=(25, 25))
    bin_statistic['statistic'] = gaussian_filter(bin_statistic['statistic'], 1)
    pcm = pitch.heatmap(bin_statistic, ax=ax, cmap='hot', edgecolors='#22312b')
    # Add the colorbar and format off-white
    cbar = fig.colorbar(pcm, ax=ax, shrink=0.6)
    cbar.outline.set_edgecolor('#efefef')
    cbar.ax.yaxis.set_tick_params(color='#efefef')
    ticks = plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='#efefef')  
    path_eff = [path_effects.Stroke(linewidth=1.5, foreground='black'),
            path_effects.Normal()]
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2, pitch_color='#22312b')
    fig, ax = pitch.draw(figsize=(9.7,8.5))
    fig.set_facecolor('#22312b')
    bin_statistic = pitch.bin_statistic(df_team_shot.x, df_team_shot.y, statistic='count', bins=(6, 5), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, cmap='Reds', edgecolor='#22312b')
    labels = pitch.label_heatmap(bin_statistic, color='#f4edf0', fontsize=18,
                             ax=ax, ha='center', va='center',
                             str_format='{:.0%}', path_effects=path_eff)
TeamShotChart('Italy')
```

```python
def TeamPasses(team):
    team_pass = (events_df.team == team) & (events_df.type == 'Pass')
    df_team_pass = events_df.loc[team_pass, ['x', 'y', 'end_x', 'end_y', 'pass_outcome']]
    team_pass_complete = df_team_pass.pass_outcome.isnull()
    #Set up the pitch
    pitch = Pitch(pitch_type='statsbomb', line_color='#000009', line_zorder=2)
    fig, ax = pitch.draw(figsize=(10, 16))
    kde = pitch.kdeplot(df_team_pass.x, df_team_pass.y, ax=ax,
                          shade=True, levels=100, shade_lowest=True,
                          cut=4, cmap='OrRd')
    ax_title = ax.set_title(f'{team} passing chart', fontsize=30, c = 'white')
    
    len_team_pass = sum((events_df.team == team) & (events_df.type == 'Pass'))
    incomplete_pass = sum((events_df.team == team) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Incomplete'))
    injcl_pass = sum((events_df.team == team) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Injury Clearance'))
    out_pass = sum((events_df.team == team) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Out'))
    offside_pass = sum((events_df.team == team) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Pass Offside'))
    other_pass = sum((events_df.team == team) & (events_df.type == 'Pass') & (events_df.pass_outcome == 'Unknown'))
    incomplete_pass_all = incomplete_pass + injcl_pass + out_pass + offside_pass + other_pass
    complete_pass = len_team_pass - incomplete_pass_all
    
    textstr = '\n'.join((
    f'Complete Passes = {round((complete_pass/len_team_pass)*100)} %',
    f'Incomplete Passes = {round((incomplete_pass_all/len_team_pass)*100) } %',
    ))
    props = dict(boxstyle='round', facecolor='#7a7d7c', alpha=0.9)
    
    plt.text(0.05, 0.95, textstr, transform= ax.transAxes, fontsize=18,
        verticalalignment='top', bbox=props)
    pearl_earring_cmap = LinearSegmentedColormap.from_list("Pearl Earring - 10 colors",
                                                       ['#15242e', '#4393c4'], N=10)
    path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]
    #Fontmanager for google font (robotto)
    robotto_regular = FontManager()
    pitch = Pitch(pitch_type='statsbomb', line_zorder=2, pitch_color='#1e4259')
    fig, axs = pitch.grid(title_height=0.08, title_space=0,
                      axis=False,
                      grid_height=0.84)
    fig.set_facecolor('#1e4259')
    #Heatmap and labels
    bin_statistic = pitch.bin_statistic_positional(df_team_pass.x, df_team_pass.y, statistic='count',
                                               positional='full', normalize=True)
    pitch.heatmap_positional(bin_statistic, ax=axs['pitch'],
                         cmap=pearl_earring_cmap, edgecolors='#22312b')
    labels = pitch.label_heatmap(bin_statistic, color='yellow', fontsize=18,
                             ax=axs['pitch'], ha='center', va='center',
                             str_format='{:.0%}', path_effects=path_eff)
    axs['title'].text(0.5, 0.5, f'{team} passing chart', color='white',
                  va='center', ha='center', path_effects=path_eff,
                  fontproperties=robotto_regular.prop, fontsize=30)
TeamPasses('Italy')
```

**With these passing charts, we are able to see that the players normally involve their goalkeeper before they begin their playmaking. They also are really present of the left side of the offensive part of the field.**
